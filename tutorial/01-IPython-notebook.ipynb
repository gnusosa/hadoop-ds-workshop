## What is it?
IPython notebooks are a browser based environment that allow you to run code interactively, view the results in place and share the results in a standard format. Recently, the IPython notebook grew into its successor, [Jupyter](https://jupyter.org), which allows for notebooks in other languages than just Python (e.g. R, Scala, Julia). Jupyter notebooks consist of the notebook server, which allows for web based access, and a kernel, which executes the code and produces results. Multiple language kernels exist. For this workshop we will focus on the Python kernel.

## How does it work?
A notebook consists of a series of cells (i.e. blocks, like this one you're reading). A cell can be one of these three:
- Markdown cell
- Code cell
- Result cell

You can change the cell type for a selected cell from the `Cell -> Cell Type` menu.

Markdown cells allow you to document the steps taken in a notebook in a structured markup language. The Markdown processor includes support for Math expressions. This cell is a Markdown cell (double click or hit Enter to see the source; hit Ctrl+Enter to see the rendered version again - Shift+Enter additionally moves focus to the next cell).

Code cells contain (Python) code and can be executed (by hitting [Ctrl|Shift]+Enter). When executed, a code cell will create a result cell directly underneath it if there is any output to display.
Some examples follow.

```python
# Python FizzBuzz
word = lambda x: ["{0}", "Fizz", "Buzz", "FizzBuzz"][(x % 3 == 0) | ((x % 5 == 0) << 1)].format(x)
num_fizzbuzz = 22
result = [word(x) for x in range(1,num_fizzbuzz)]

# The final expression of a cell becomes its output. No need for print statements.
result
```

# Scoping, variables and functions

Everything that you define in a notebook is scoped globally. When you assign a variable in one cell, it is available in all subsequent cells. The same goes for functions.

```python
# E.g. the 'result' variable that was assigned earlier, can still be used here
result[14]
```

```python
# Define a function
def square(x):
    return x ** 2

# This cell doesn't generate any output, so there's no output cell.
```

```python
# Use a previously defined function
square(9)
```

## Imports and reusability

In Python you can import from packages that are installed in the environment, and you can import Python modules that exist in the current directory. In Python, each source file is a module.

As a rule of thumb, it makes sense to create a _module_ for code that you reuse across multiple notebooks focusing on the same problem, while you would create a _package_ for code that you share more broadly and solves a generic problem. For example, we will use a simple _module_ that contains functions for processing the data that is specific to this workshop, and we will use third party _packages_ for plotting and data analysis.

```python
# Imports can appear anywhere in the notebook
from example_module import greetings

# You can review the module's source code by opening the example_module.py
# from the same directory. This can be done from the web interface as well,
# Jupyter notebook embeds a web based editor with highlighting for many
# languages.

# We've imported this function
greetings()
```

## Saving and sharing notebooks

Notebooks are saved as .ipynb files that contain a JSON structure of all notebook data. When saved, not only the Markdown and code are saved, but also the result cells, including any images, HTML and other output. This means that it's possible to render a notebook, without being able to execute the code. This makes notebooks great for sharing research and other snippets. Additionally, notebooks are rendered by Github these days, so you can view this notebook online by clicking [here](https://github.com/godatadriven/hadoop-ds-workshop/blob/master/tutorial/01-IPython-notebook.ipynb).
